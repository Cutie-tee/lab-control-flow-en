**How many people have kids**?

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]

count_kids = 0
# Iterate over each person in the list
for person in people:
    if person["n_kids"] > 0:  # Check if the person has kids
        count_kids += 1  # Increment the counter

# Print the final count
print("Number of people with kids:", count_kids)
```

**How many kids do they have in total?**

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]

#  counter for the total number of kids
total_kids = 0
# Iterate over each person in the list
for person in people:
    total_kids += person["n_kids"]  # Add the number of kids to the total

# Print the total number of kids
print("Total number of kids:", total_kids)
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
# Create a list to store the updated information
people_in_a_year = []

# Go through each person in the original list
for person in people:
    # If the person's name ends with 'a', increase the number of kids by 1
    if person["name"].endswith("a"):
        person["n_kids"] += 1
    
    # Add the person (with updated info if needed) to the new list
    people_in_a_year.append(person)

# Print the updated list
print(people_in_a_year)
```
